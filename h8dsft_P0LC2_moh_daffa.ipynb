## Problem 1-1 What cities are included in the distribution area coverage?

```python
# untuk kita dapat mencari are yang tercakup distribusi maka kita harus mencari unique value atau mencari kota apa saja yang terdapat dalam list.
```

```python
#untuk ini kita menggunakan select distinct, untuk mencari kota-kota yang terdapat di dalam list data
df_1 = client.query('''
              SELECT DISTINCT store_city AS included_distribution_area_coverage
              FROM `ftds-hacktiv8-project.phase0_ftds_003_bsd.beverage_sales`
              ''').to_dataframe()

df_1
```

## Problem 1-2 What is the minimum and maximum range for frinking units sold?

```python
# untuk mencari min dan max dan avg kita dapa menggunakan min/max/avg dengan unit_sold kita karena kita mencari harga terendah dan harga tertinggi didalam data
df_2 = client.query('''
              SELECT MIN(unit_sold) AS min_unit_sold,
                    MAX(unit_sold) AS max_unit_sold,
                    AVG(unit_sold) AS mean_unit_sold
              FROM `ftds-hacktiv8-project.phase0_ftds_003_bsd.beverage_sales`
              ''').to_dataframe()
# kita menggunakan df_2 untuk calling function karena kita telah membuat function kita dengan df_2
df_2
```

## Problem 1-3 Compare total sales and number of transactions from each city in 2022

```python
# Problem 1-3 Compare total sales and number of transactions from each city in 2022
```

```python
# untuk mencari total sales dan number of transactions kita menggunakan extract(year from) karena kita mencari berapa total tersebut
# dilanjutkan dengan count(unit_sold
df_3 = client.query('''
              SELECT EXTRACT(YEAR FROM date) AS years,
                    store_city,
                    COUNT(unit_sold) AS total_sales
              FROM `ftds-hacktiv8-project.phase0_ftds_003_bsd.beverage_sales`
              WHERE EXTRACT(YEAR FROM date) = 2022
              GROUP BY 1,2
              ORDER BY 3 DESC
              ''').to_dataframe()

df_3.head()
```

## Problem 1-4 What products will be most in demand in 2022
